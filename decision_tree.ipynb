```python
import pandas as pd

# สร้าง DataFrame
data = pd.DataFrame({
    'face':[0, 0, 0, 1, 1, 0],
    'ear':[1, 0, 0, 1, 1, 1],
    'brown':[1, 1, 1, 0, 1, 0],
    'tongue':[1, 1, 0, 0, 0, 0],
    'whisker':[1, 0, 0, 1, 1, 1],
    'target':['dog', 'dog', 'dog', 'cat', 'cat', 'cat']
})
data
```

```python
data['target'] = data['target'].map({'cat':0, 'dog':1})
```

```python
import numpy as np

# แยก Features (X)
X = data[['face','ear','brown','tongue','whisker']].values
# แยก Target (y)
y = data['target'].values
print("X =\n", X)
print("y =\n", y)
print("Shape X:", X.shape)
print("Shape y:", y.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

# สร้างโมเดลต้นไม้
model = DecisionTreeClassifier(random_state=42)
# Train โมเดล
model.fit(X, y)
# ทดสอบความถูกต้อง
accuracy = model.score(X, y)
print("Accuracy:", accuracy)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(10,6))
plot_tree(
    model,
    feature_names=['face','ear','brown','tongue','whisker'],
    class_names=['cat','dog'],
    filled=True,
    rounded=True
)
plt.title("Decision Tree: Cat vs Dog")
plt.show()
```
